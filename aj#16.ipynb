```python
nitrogen = int(input("enter nitrogen level"))
print(f"soil nitrogen level: {nitrogen} mg/kg")
if nitrogen < 100:
    print("apply high dose fertiliser,\ncost=$2000/bag,\n message=soil fertility is poor")
elif 100 <= nitrogen < 250:
    print("apply medium dose fertiliser,\ncost=$1000/bag,\n message=soil fertility is average")
else:
    print("apply low dose fertiliser,\ncost=$500/bag,\n message=soil fertility is good")
```
