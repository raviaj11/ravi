```python
def drop_lowest(marks):
  new_marks=marks.copy()
  new_marks.remove(min(new_marks))
  return new_marks

def average(marks):
  total=0
  for m in marks:
    total=total+m
  return total/len(marks)

def is_passing(avg,threshould):
  #return true or False
  if avg>= threshould:
    return True
  else:
    return False

#Main flow
marks_list=[45,60,75,30]
threshould=50
new_marks=drop_lowest(marks_list)
avg=average(new_marks)
result=is_passing(avg,threshould)
print("Marks after dropping lowest:",new_marks)
print("Average:",avg)
print("pass?",result)
```

```python
student = [] # Initialize student list
# Add 03 students
for i in range(3):
    roll = int(input("Enter roll:"))
    name = input("Enter name:")
    m1 = int(input("Enter marks1:"))
    m2 = int(input("Enter marks2:"))
    m3 = int(input("Enter marks3:"))
    record = {
        "roll": roll,
        "name": name,
        "marks": [m1, m2, m3]
    }
    student.append(record)

# Print each student name and average mark
for s in student:
    marks = s["marks"]
    avg = sum(marks) / len(marks)
    print(s["roll"], s["name"], "Average:", avg)
```
